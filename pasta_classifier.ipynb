## Getting the data

```python
import os
from pathlib import Path
from PIL import Image
import shutil
```

```python
# Downloading online images 
from simple_image_download import simple_image_download as simp 
response = simp.simple_image_download
lst=[
    'spaghetti', 
    'fusilli',
    'conchiglioni',
    'farfalle',
    'fettuccine',
    'tagliatelle',
    'gnocchi',
    'lasagne',
    'linguine',
    'penne',
    'orecchiette',
    'rigatoni', 
]
for rep in lst:
 response().download(rep + " raw pasta", 2000)
```

```python
def cleaning_images(directory_path):
    if (directory_path / '.ipynb_checkpoints').is_dir():
        shutil.rmtree(directory_path / '.ipynb_checkpoints')
    num_of_images = 0
    num_of_deleted_images = 0
    for file_path in directory_path.iterdir():
        try:
            im = Image.open(file_path)
        except:
            os.remove(file_path)
            num_of_deleted_images +=1 
            
    for file_path in directory_path.iterdir():
        im = Image.open(file_path)
        if im.size[0] > 320 and im.size[1] > 320:
            num_of_images += 1
        else:
            os.remove(file_path)
            num_of_deleted_images +=1 

    return num_of_images, num_of_deleted_images
```

```python
cleaning_images(Path("simple_images/conchiglioni raw pasta/"))
```

```python
simple_images = Path('simple_images/')

for dir_path in simple_images.iterdir():
    num_of_images, nu_of_deleted_images = cleaning_images(dir_path)
    print(dir_path, num_of_images, nu_of_deleted_images)
```
